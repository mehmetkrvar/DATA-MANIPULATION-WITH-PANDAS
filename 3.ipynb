Indexes and subsetting using indexes

```python
import pandas as pd
dogs = pd.read_csv('dogs.csv')
dogs.head()
```

```python
#Setting a column as the index
dogs_index =dogs.set_index('name')
print(dogs_index)
```

```python
#Removing an index
dogs_index.reset_index()
```

```python
#Dropping an index
dogs_index.reset_index(drop=True)
```

```python
#Indexes make subsetting simpler
dogs[dogs['name'].isin(['Bella','Stella'])]
```

```python
dogs_index.loc[['Bella','Stella']]
```

```python
#Index values don't need to be unique
dogs_index2 = dogs.set_index('breed')
print(dogs_index2)
```

```python
#Subsetting on duplicated index values
dogs_index2.loc['Labrador']
```

```python
#Multi-level indexes a.k.a. hierarchical indexes
dogs_index3 = dogs.set_index(['breed','color'])
print(dogs_index3)
```

```python
#Subset the outer level with a list
dogs_index3.loc[["Labrador", "Chihuahua"]]
```

```python
#Subset inner levels with a list of tuples
dogs_index3.loc[[('Labrador','Brown'), ( 'Chihuahua','Tan')]]
```

```python
#Sorting by index values
dogs_index3.sort_index()
```

```python
#Controlling sort_index
dogs_index3.sort_index(level=['color','breed'],ascending=[True,False])
```

```python
#Sort the index before you slice
dogs_srt = dogs.set_index(['breed','color']).sort_index()
dogs_srt
```

```python
#Slicing the outer index level

dogs_srt.loc['Chow Chow' : 'Poodle']
```

```python
#Slicing the inner index levels badly
dogs_srt.loc['Tan': 'Grey']
```

```python
#Slicing the inner index levels correctly
dogs_srt.loc[('Labrador', 'Brown') : ('Schnauzer', 'Grey')]
```

```python
#Slicing columns
dogs_srt.loc[:,'name' : 'height_cm']
```

```python
#Slice twice
dogs_srt.loc[('Labrador','Brown') : ('Schnauzer','Grey'), 'name' : 'height_cm']
```

```python
#Dog days
dogs = dogs.set_index('date_of_birth').sort_index()
dogs
```

```python
#Slicing by dates
dogs.loc['2014-08-25':'2016-09-16']
```

```python
#Slicing by partial dates
dogs.loc['2014':'2016']
```

```python
#Subsetting by row/column number
print(dogs.iloc[2:5,1:4])
```

```python
#A bigger dog dataset
dog_pack = pd.read_csv('dogs_pack.csv')
dog_pack.head()
```

```python
#Pivoting the dog pack
aaa= dog_pack.pivot_table(values='height_cm',index='breed',columns='color')
print(aaa)
```

```python
#.loc[] + slicing is a power combo
aaa.loc['Chow Chow' : 'Poodle']
```

```python
#The axis argument
aaa.mean(axis='index')
```

```python
#Calculating summary stats across columns
aaa.mean(axis='columns')
```
