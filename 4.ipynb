```python
import pandas as pd
dog_pack = pd.read_csv('dogs_pack.csv')
dog_pack.head()
```

```python
#Histograms
import matplotlib.pyplot as plt
dog_pack['height_cm'].hist()
plt.show()
```

```python
#Histograms
dog_pack['height_cm'].hist(bins=20)
plt.show()
```

```python
#Bar plots
avg_weight = dog_pack.groupby('breed')['weight_kg'].mean()
print(avg_weight)
```

```python
avg_weight.plot(kind='bar')
plt.show()
```

```python
avg_weight.plot(kind='bar',title='mean wight by dog breed')
plt.show()
```
